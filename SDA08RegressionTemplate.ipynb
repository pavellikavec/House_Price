# Preprocessing

```python
y = df['SalePrice']

new_submission = df_kaggle[["Id"]] #Uložení Id bokem
new_submission.head()
```

```python
#zpracovává jak df, tak df_kaggle, doporučení je rozšiřovat tuto funkci o preprocessingové parametry tak, aby oba dataframes měly stejný shape
def preproces_data(x):
  x_output = x.select_dtypes(exclude=['object']) #vyhození všech proměnných typu object
  x_output = x_output.fillna(0) #do všech nedefinovaných hodnot naplní 0
  for var in ['Id', 'SalePrice']: 
    if var in x_output.columns:
      x_output.drop(var, inplace=True, axis =1) #dropne se Id a Saleprice pokud je ve sloupcích
  return(x_output)
```

```python
X = preproces_data(df)
X_kaggle = preproces_data(df_kaggle)
```

# Modelling

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
m = LinearRegression()
m.fit(X_train,y_train)
```

```python
y_pred = m.predict(X_test)

# if prediction is negative:
# mean_squared_error(np.log(np.where(ypred<=0, np.mean(y_train), ypred)), np.log(y_test), squared = False)

mean_squared_error(np.log(y_pred), np.log(y_test), squared = False) #pokud vyjde záporně, může spadnout na chybu
```

# Submission To Kaggle

```python
y_pred = m.predict(X_kaggle)
#případně fit na všech datech m.fit(X, y)
new_submission['SalePrice'] = y_pred
```

# Variable importance

```python
errors = []
for var in X_train.columns:
  X_train_drop = X_train.drop([var], axis = 1)
  X_test_drop = X_test.drop([var], axis = 1)
  m = LinearRegression()
  m.fit(X_train_drop,y_train)
  y_pred = m.predict(X_test_drop)
  errors.append(mean_squared_error(np.log(y_pred), np.log(y_test), squared = False))
```

```python
errors_np = np.array(errors)
errors_np
```
